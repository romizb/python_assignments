```python
import pandas as pd

# Load the dataset
data = pd.read_csv('patient_data.csv')  # the vsc file we open

print("First 5 rows of the dataset:") 
print(data.head())  #Shows the first 5 rows of the dataset by default, if want to show more, write data.head(10)...

num_patients = len(data) #Returns the number of rows in the DataFrame
print(f"\nTotal number of patients in the dataset: {num_patients}")
```

```python
# Calculate basic statistics

mean_wbc = data['WBC_Count'].mean() # Compute the mean WBC count

min_wbc = data['WBC_Count'].min() # find the minimum WBC count

max_wbc = data['WBC_Count'].max() # find the maximum WBC count

print(f"\nMean WBC count: {mean_wbc}")
print(f"Minimum WBC count: {min_wbc}")
print(f"Maximum WBC count: {max_wbc}")

unique_conditions = data['Condition'].nunique() # Count "unique conditions"
print(f"\nNumber of unique conditions: {unique_conditions}")
```

```python
# Filter patients with WBC count > 10,000
high_wbc = data[data['WBC_Count'] > 10000]  
print("\nPatients with WBC count above 10,000:")
print(high_wbc)

# Filter patients with the condition 'Healthy'
healthy_patients = data[data['Condition'] == 'Healthy']  
print("\nPatients with the condition 'Healthy':")
print(healthy_patients)
```

```python
# Counts the occurrences of each unique value in the 'Condition' column
condition_counts = data['Condition'].value_counts()
print(condition_counts)
```

```python
#visualise data
import matplotlib.pyplot as plt

# Bar chart for number of patients by condition
plt.figure(figsize=(8, 5))
condition_counts.plot(kind='bar', color='skyblue', edgecolor='black')  # Creates a bar chart
plt.title('Number of Patients by Condition')
plt.xlabel('Condition')
plt.ylabel('Number of Patients')
plt.xticks(rotation=45)  # Rotates the x-axis labels for better readability
plt.tight_layout()  # Adjusts layout to prevent clipping
plt.show()

# Scatter plot of Age vs. WBC Count
plt.figure(figsize=(8, 5))
plt.scatter(data['Age'], data['WBC_Count'], color='darkorange', edgecolor='black', alpha=0.8)
plt.title('Age vs. WBC Count')
plt.xlabel('Age')
plt.ylabel('WBC Count')
plt.grid(True, linestyle='--', alpha=0.7)  # Adds a grid to the plot
plt.tight_layout()
plt.show()
```
